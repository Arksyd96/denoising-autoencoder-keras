## **Imports:**

```python
import cv2
import keras.backend as K
from keras.layers import Conv2D
from keras.layers import Conv2DTranspose
from keras.layers import BatchNormalization
from keras.layers import Input
from keras.layers import Dense
from keras.layers import Flatten
from keras.models import Model
from keras.layers import Reshape
from keras.datasets import cifar10, mnist
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np

print('imported')
```

# **Creating the encoder:**
#### Three convolution layers followed by a batch normalization. ReLU as activation function (we could use SReLU for more performance)
#### We fix the latent dimension to 1024, the more the latentDim is big, the more it can contain informations and the decoder will be able to reconstruct the image better

```python
latentDim = 1024

encoderInput = Input(shape=(32, 32, 3))
X = Conv2D(32, (3, 3), strides=2, padding="same", activation="relu")(encoderInput)
X = BatchNormalization(axis=-1)(X)
X = Conv2D(64, (3, 3), strides=2, padding="same", activation="relu")(X)
X = BatchNormalization(axis=-1)(X)
X = Conv2D(128, (3, 3), strides=2, padding="same", activation="relu")(X)
X = BatchNormalization(axis=-1)(X)
volumeSize = K.int_shape(X)
X = Flatten()(X)
X = Dense(latentDim)(X)

encoder = Model(encoderInput, X, name="encoder")
encoder.summary()
```

# **creating the decoder**:
#### We Reshape to get the initial 3 dimensions of an image.
#### Conv2DTranspose does a deconvolution and as always followed by a Batch normalization
#### Output activation is a sigmoid because of the normalized image data

```python
decoderInput = Input(shape=(latentDim,))
X = Dense(np.prod(volumeSize[1:]))(decoderInput)
X = Reshape((volumeSize[1], volumeSize[2], volumeSize[3]))(X)
X = Conv2DTranspose(128, (3, 3), strides=2, padding="same", activation="relu")(X)
X = BatchNormalization(axis=-1)(X)
X = Conv2DTranspose(64, (3, 3), strides=2, padding="same", activation="relu")(X)
X = BatchNormalization(axis=-1)(X)
X = Conv2DTranspose(32, (3, 3), strides=2, padding="same", activation="relu")(X)
X = BatchNormalization(axis=-1)(X)
X = Conv2DTranspose(3, (3, 3), padding="same", activation="sigmoid")(X)

decoder = Model(decoderInput, X, name="decoder")
decoder.summary()
```

# **Autoencoder:**

```python
autoencoder = Model(encoderInput, decoder(encoder(encoderInput)), name="autoencoder")
autoencoder.summary()
```

# **Preparing data (CIFAR10 dataset):**
#### Loading dataset from Keras.datasets and applying a noise effect, this consists of adding random small values to our image

```python
(x_train, y_train), (x_test, y_test) = cifar10.load_data()
x_train = x_train.astype("float32") / 255.0
y_train = y_train.astype("float32") / 255.0
x_test = x_test.astype("float32") / 255.0
y_test = y_test.astype("float32") / 255.0

trainNoise = np.random.normal(size=x_train.shape) * (30 / 255)
testNoise = np.random.normal(size=x_test.shape) * (30 / 255)
x_train_noisy = np.clip(x_train + trainNoise, 0, 1)
x_test_noisy = np.clip(x_test + testNoise, 0, 1)
```

# **Training our model:**
#### Adamax as an optimizer, i prefer Adamax to Adam
#### We can also use MAE loss function instead of MSE, give it a try it can give better results

```python
autoencoder.compile(loss="mse", optimizer="adamax")

history = autoencoder.fit(x_train_noisy, x_train,
               validation_data=(x_test_noisy, x_test),
               epochs=20,
               batch_size=32)
```

```python
autoencoder.fit(x_train, x_train,
               validation_data=(x_test, x_test),
               epochs=5,
               batch_size=32)
```

## **Ploting loss and accuracy:**

```python
N = np.arange(0, 20)
plt.style.use("ggplot")
plt.figure()
plt.plot(N, history.history["loss"], label="train_loss")
plt.plot(N, history.history["val_loss"], label="val_loss")
plt.title("Training Loss and Accuracy")
plt.xlabel("Epoch #")
plt.ylabel("Loss/Accuracy")
plt.legend(loc="lower left")
```

## **Some examples:**

```python
decoded = autoencoder.predict(x_test_noisy)

plt.figure(figsize=(20,20))
for i in range(0, 10):
    original = (x_test[i] * 255).astype("uint8")
    noisy = (x_test_noisy[i] * 255).astype("uint8")
    recon = (decoded[i] * 255).astype("uint8")
    plt.subplot(10, 1, i + 1)
    plt.imshow(np.hstack([original, noisy, recon]))
    plt.axis("off")
plt.show()
```

## **Saving our model to disk:**

```python
# serialize model to JSON
model_json = autoencoder.to_json()
with open("autoencoder.json", "w") as json_file:
    json_file.write(model_json)
# serialize weights to HDF5
autoencoder.save_weights("autoencoder.h5")
print("Saved model to disk")
```
