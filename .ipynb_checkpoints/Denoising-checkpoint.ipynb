```python
import cv2
import keras.backend as K
from keras.layers import Conv2D
from keras.layers import Conv2DTranspose
from keras.layers import BatchNormalization
from keras.layers import Input
from keras.layers import Dense
from keras.layers import Flatten
from keras.models import Model
from keras.layers import Reshape
from keras.datasets import cifar10, mnist
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

print('imported')
```

```python
x = Input(shape=(32, 32, 3))
```

**Creating the encoder:**

```python
latentDim = 128

encoderInput = Input(shape=(32, 32, 3))
X = Conv2D(32, (3, 3), strides=2, padding="same", activation="relu")(encoderInput)
X = BatchNormalization(axis=-1)(X)
X = Conv2D(64, (3, 3), strides=2, padding="same", activation="relu")(X)
X = BatchNormalization(axis=-1)(X)
X = Conv2D(64, (3, 3), strides=1, padding="same", activation="relu")(X)
X = BatchNormalization(axis=-1)(X)
X = Conv2D(128, (3, 3), strides=2, padding="same", activation="relu")(X)
X = BatchNormalization(axis=-1)(X)
X = Conv2D(256, (3, 3), strides=2, padding="same", activation="relu")(X)
X = BatchNormalization(axis=-1)(X)
volumeSize = K.int_shape(X)
X = Flatten()(X)
X = Dense(latentDim)(X)

encoder = Model(encoderInput, X, name="encoder")
encoder.summary()
```

**Creating the decoder:**

```python
decoderInput = Input(shape=(latentDim,))
X = Dense(np.prod(volumeSize[1:]))(decoderInput)
X = Reshape((volumeSize[1], volumeSize[2], volumeSize[3]))(X)
X = Conv2DTranspose(256, (3, 3), strides=2, padding="same", activation="relu")(X)
X = BatchNormalization(axis=-1)(X)
X = Conv2DTranspose(128, (3, 3), strides=2, padding="same", activation="relu")(X)
X = BatchNormalization(axis=-1)(X)
X = Conv2DTranspose(64, (3, 3), strides=1, padding="same", activation="relu")(X)
X = BatchNormalization(axis=-1)(X)
X = Conv2DTranspose(64, (3, 3), strides=2, padding="same", activation="relu")(X)
X = BatchNormalization(axis=-1)(X)
X = Conv2DTranspose(32, (3, 3), strides=2, padding="same", activation="relu")(X)
X = BatchNormalization(axis=-1)(X)
X = Conv2DTranspose(3, (3, 3), padding="same", activation="sigmoid")(X)

decoder = Model(decoderInput, X, name="decoder")
decoder.summary()
```

**Autoencoder:**

```python
autoencoder = Model(encoderInput, decoder(encoder(encoderInput)), name="autoencoder")
autoencoder.summary()
```

**Preparing data: (MNIST)**

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train = x_train.astype("float32") / 255.0
y_train = y_train.astype("float32") / 255.0
x_test = x_test.astype("float32") / 255.0
y_test = y_test.astype("float32") / 255.0

trainNoise = np.random.normal(loc=0.5, scale=0.5, size=x_train.shape)
testNoise = np.random.normal(loc=0.5, scale=0.5, size=x_test.shape)
x_train_noisy = np.clip(trainNoise + x_train, 0, 1)
x_test_noisy = np.clip(testNoise + x_test, 0, 1)

x_train_noisy = np.expand_dims(x_train_noisy, axis=-1)
x_test_noisy = np.expand_dims(x_test_noisy, axis=-1)
x_train = np.expand_dims(x_train, axis=-1)
x_test = np.expand_dims(x_test, axis=-1)
```

**Preparing data: (Cifar10)**

```python
(x_train, y_train), (x_test, y_test) = cifar10.load_data()
x_train = x_train.astype("float32") / 255.0
y_train = y_train.astype("float32") / 255.0
x_test = x_test.astype("float32") / 255.0
y_test = y_test.astype("float32") / 255.0

trainNoise = np.random.normal(size=x_train.shape) * (30 / 255)
testNoise = np.random.normal(size=x_test.shape) * (30 / 255)
x_train_noisy = np.clip(x_train + trainNoise, 0, 1)
x_test_noisy = np.clip(x_test + testNoise, 0, 1)
```

```python
autoencoder.compile(loss="mae", optimizer="adam")

history = autoencoder.fit(x_train_noisy, x_train,
               validation_data=(x_test_noisy, x_test),
               epochs=25,
               batch_size=32)
```

```python
N = np.arange(0, 25)
plt.style.use("ggplot")
plt.figure()
plt.plot(N, history.history["loss"], label="train_loss")
plt.plot(N, history.history["val_loss"], label="val_loss")
plt.title("Training Loss and Accuracy")
plt.xlabel("Epoch #")
plt.ylabel("Loss/Accuracy")
plt.legend(loc="lower left")
```

```python
decoded = autoencoder.predict(x_test_noisy)
outputs = None

for i in range(0, 10):
    #origianl = (x_test[i] * 255).astype("uint8")
    noisy = (x_test_noisy[i] * 255).astype("uint8")
    recon = (decoded[i] * 255).astype("uint8")

    output = np.hstack([noisy, recon])

    if outputs is None:
        outputs = output
    else:
        outputs = np.vstack([outputs, output])
cv2.imwrite("output.jpg", outputs)
```

```python
plt.figure(figsize=(20,20))
for i in range(0, 10):
    original = (x_test[i] * 255).astype("uint8")
    noisy = (x_test_noisy[i] * 255).astype("uint8")
    recon = (decoded[i] * 255).astype("uint8")
    plt.subplot(10, 1, i + 1)
    plt.imshow(np.hstack([original, noisy, recon]))
    plt.axis("off")
plt.show()
```
